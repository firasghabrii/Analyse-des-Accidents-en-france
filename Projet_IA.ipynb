```python
tableau_participants = soup.find('table', {'class' : 'wikitable'})
tableau_participants
```

```python
table_body = tableau_participants.find('tbody')
rows = table_body.find_all('tr')
print(rows[0].text.strip())
```

```python
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    print(cols)
```

```python
dico_participants = dict()
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    if len(cols) > 0 : 
        dico_participants[cols[0]] = cols[1:]
dico_participants
```

```python
data_accident = pd.DataFrame.from_dict(dico_participants,orient='index')
data_accident.head(71)
```

```python
data_accident.columns = ['Tués','Accidents']
data_accident.index.name="Département"
data_accident.head(71)
```
